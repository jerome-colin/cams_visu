```python
import xarray as xr
import pandas as pd
import matplotlib.pylab as pl
from datetime import datetime
import altair as alt
```

```python
lat = 22.63376
lon = 88.37364
site = "calculta"
```

```python
ds = xr.open_dataset('/home/colinj/data/cams/stacked_aot/stacked_cams_aot.nc')
df_calcuta = ds.sel(latitude=lat, longitude=lon, method="nearest").to_dataframe()
df_calcuta.pop('latitude')
df_calcuta.pop('longitude')
df_calcuta = df_calcuta.sort_index()
df_calcuta.reset_index(inplace=True)
df_calcuta.describe()
```

```python
domain = ['amaod550', 'bcaod550', 'duaod550', 'niaod550', 'omaod550', 'ssaod550', 'suaod550']
range_ = ['olive', 'grey', 'orange', 'cyan', 'green', 'blue', 'red']

alt.Chart(df_calcuta).transform_fold(
    ['duaod550',
     'omaod550',
     'bcaod550',
     'suaod550',
     'niaod550',
     'ssaod550',
     'amaod550'],
    as_ = ['Aerosols', 'AOD']
).mark_area().encode(
    alt.X('time:T', axis=alt.Axis(format='%y/%m/%d')),
    alt.Y('AOD:Q', stack='zero'),
    alt.Color('Aerosols:N', scale=alt.Scale(domain=domain, range=range_))
    #alt.Color('Aerosols:N')
).properties(
    width=1200,
    height=400,
    title='CAMS aerosols over Calcuta'
).interactive()
```
